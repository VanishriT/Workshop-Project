```python
#Data visualization
#1)How has startup funding trend changed over the years?
import matplotlib.pyplot as plt
demo = data.groupby("year")["amount"].sum()
plt.figure(figsize=(10,6))
plt.plot(demo.index,demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.xlabel("Year",color="blue",fontsize=12)

plt.ylabel("Amount",color="blue",fontsize=12)
plt.title("Year wise amount investment ",color="red",fontsize=16)
plt.grid(linestyle=":",color="skyblue",linewidth=1.5)
plt.legend()
plt.show()
```

```python
#2) Which cities have the highest number of startup

demo=data["city"].value_counts().head(10)
colors=["#FA6781","#934761","#E13F7C","#427AB5","#F08D39","#D62828","#D96868","#355872","#F0FFC3","#0C7779"]

plt.figure(figsize=(10,6))
plt.bar(demo.index,demo.values,color=colors,edgecolor="grey")
plt.xticks(rotation = 45)
plt.title("Top 10 startup comapny's in India")
plt.xlabel("City Names",color="#005461",fontsize=12)
plt.ylabel("Number of startup",color="#005461",fontsize=12)
plt.grid(True)
plt.show()
```

```python
#3) Which startup sectors have the highest number of funded startup?
demo=data["vertical"].value_counts().head(10)
colors=["#499A13","#BD4444","#00E0BA","#458393","#293681","#443199","#6FCF97","#AD5C71","#AE2448","#2C687B"]

plt.figure(figsize=(10,7))
plt.barh(demo.index[::-1],demo.values[::-1],color=colors,edgecolor="grey",alpha=0.9)
plt.xticks(rotation = 45)
plt.title("Top 10 Sectors startups in India",fontsize=20)
plt.xlabel("Number of startups",color="#005461",fontsize=12)
plt.ylabel(" Sectors",color="#005461",fontsize=12)
plt.grid(axis='x',linestyle="--",color="#458B73")
plt.show()
```

```python
#4) What is the distribution of different funding rounds?
import matplotlib.pyplot as plt

demo=data["round"].value_counts().head(7)
colors=["#499A13","#F5824A","#088395","#FF0087","#0B2D72","#00F7FF","#5D2F77"]
plt.figure(figsize=(10,6))

plt.figure(figsize=(10,7))
plt.pie(demo,labels=demo.index,colors=colors,autopct="%1.1f%%")
plt.title("Rounds distribution in percentage",color="#831C91",fontsize=15)
plt.show()

```

```python
#5) How are startup funding amounts distributed?

plt.hist(data["amount"], bins=10, edgecolor="black")
plt.title("Distribution of startup funding",color="#088395",fontsize=15)
plt.show()
```

```python
#6) The Relationship between Funding year and Funding amount?

X=data[['year']]
y=data['amount']
plt.figure(figsize=(10,6))
plt.scatter(X,y,color="#DE1A58")
plt.grid(True,linestyle="--")
plt.title("Funding amount by year",color="#1A05A2",fontsize=15)
plt.xlabel("Funding Year",color="#1B4EF5",fontsize=12)
plt.ylabel("Funding Amount",color="#1B4EF5",fontsize=12)

```

```python
#8) Which funding round occurs most frequently?
import seaborn as sns

Demo = data["round"].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(data,y="round",order = Demo,palette="viridis" )
plt.title("Most frequently funding round")

```
